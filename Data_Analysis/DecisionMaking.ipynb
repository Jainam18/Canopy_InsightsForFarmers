```python
import random
import math

def generate_groundwater_data():
    data = {}
    base_level = 100  # Base groundwater level in centimeters
    
    for week in range(1, 53):
        week_key = f'week{week}'
        
        # Seasonal variation (sine wave with annual period)
        seasonal_var = 15 * math.sin(2 * math.pi * week / 52)
        
        # Calculate mean groundwater level
        mean_level = base_level + seasonal_var
        
        # Add some variation to represent aggregated data over 10 years
        variation = random.uniform(-10, 10)
        mean_level += variation
        
        # Ensure non-negative values
        mean_level = max(0, mean_level)
        
        # Generate min and max values
        # These represent the extreme values observed over the 10-year period
        min_level = max(0, mean_level - random.uniform(10, 25))
        max_level = mean_level + random.uniform(10, 25)
        
        data[week_key] = {
            'max': round(max_level, 2),
            'mean': round(mean_level, 2),
            'min': round(min_level, 2)
        }
    
    return data

# Generate the data
groundwater_data = generate_groundwater_data()

# Print first and last week as a sample
print("First week:", groundwater_data['week1'])
print("Last week:", groundwater_data['week52'])
```

```python
import random
import math

def generate_runoff_data():
    data = {}
    base_level = 50  # Base runoff level in millimeters
    
    for week in range(1, 53):
        week_key = f'week{week}'
        
        # Seasonal variation (e.g., more runoff during rainy seasons)
        seasonal_var = 20 * math.sin(2 * math.pi * week / 52)
        
        # Calculate mean runoff level
        mean_level = base_level + seasonal_var
        
        # Add some random variation to represent aggregated data over 10 years
        variation = random.uniform(-15, 15)
        mean_level += variation
        
        # Ensure non-negative values
        mean_level = max(0, mean_level)
        
        # Generate min and max values
        min_level = max(0, mean_level - random.uniform(15, 30))
        max_level = mean_level + random.uniform(15, 30)
        
        data[week_key] = {
            'max': round(max_level, 2),
            'mean': round(mean_level, 2),
            'min': round(min_level, 2)
        }
    
    return data

# Generate the data
runoff_data = generate_runoff_data()

# Print first and last week as a sample
print("First week (Runoff):", runoff_data['week1'])
print("Last week (Runoff):", runoff_data['week52'])
```

```python
def generate_flood_data():
    data = {}
    base_flood_risk = 10  # Base flood risk percentage
    
    for week in range(1, 53):
        week_key = f'week{week}'
        
        # Seasonal variation (higher flood risk during certain seasons)
        seasonal_var = 25 * math.sin(2 * math.pi * week / 52)
        
        # Calculate mean flood risk
        mean_risk = base_flood_risk + seasonal_var
        
        # Add random variation to represent aggregated data over 10 years
        variation = random.uniform(-5, 5)
        mean_risk += variation
        
        # Ensure non-negative values and cap at 100%
        mean_risk = max(0, min(100, mean_risk))
        
        # Generate min and max risk values
        min_risk = max(0, mean_risk - random.uniform(5, 15))
        max_risk = min(100, mean_risk + random.uniform(5, 15))
        
        data[week_key] = {
            'max': round(max_risk, 2),
            'mean': round(mean_risk, 2),
            'min': round(min_risk, 2)
        }
    
    return data

# Generate the data
flood_data = generate_flood_data()

# Print first and last week as a sample
print("First week (Flood):", flood_data['week1'])
print("Last week (Flood):", flood_data['week52'])
```

```python
def generate_temperature_data():
    data = {}
    base_temp = 20  # Base temperature in degrees Celsius
    
    for week in range(1, 53):
        week_key = f'week{week}'
        
        # Seasonal variation (warmer in summer, colder in winter)
        seasonal_var = 10 * math.sin(2 * math.pi * week / 52)
        
        # Calculate mean temperature
        mean_temp = base_temp + seasonal_var
        
        # Add random variation to represent aggregated data over 10 years
        variation = random.uniform(-5, 5)
        mean_temp += variation
        
        # Generate min and max temperatures
        min_temp = mean_temp - random.uniform(5, 10)
        max_temp = mean_temp + random.uniform(5, 10)
        
        data[week_key] = {
            'max': round(max_temp, 2),
            'mean': round(mean_temp, 2),
            'min': round(min_temp, 2)
        }
    
    return data

# Generate the data
temperature_data = generate_temperature_data()

# Print first and last week as a sample
print("First week (Temperature):", temperature_data['week1'])
print("Last week (Temperature):", temperature_data['week52'])
```
